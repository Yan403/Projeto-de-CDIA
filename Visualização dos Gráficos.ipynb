GRÁFICO DE BARRA 


Com o gráfico de barra, pode-se agrupar os dados pelas categorias dos

jogos e ordena-los pela classificação 5 estrelas, com isso, pode-se aferir que

o gênero de jogos mais bem avaliado da Play Store é o gênero de ação.

GRÁFICO DE DISPERSÃO

Utilizando um gráfico de dispersão, uma análise pode ser feita questionando-se se o 

número de avaliações 5 estrelas se relaciona de alguma forma com as instalações

dos jogos. Verifica-se uma relativa correlação, pois à medida que o número de notas

5 estrelas caem, o número de instalações também caem
